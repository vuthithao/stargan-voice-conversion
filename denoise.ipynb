```python
import IPython
from scipy.io import wavfile
import scipy.signal
import numpy as np
import matplotlib.pyplot as plt
import librosa
%matplotlib inline

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
# Giong nu mien bac - xin

wav_loc = "samples/cpsyn-p0_1006_8.wav"
rate, data = wavfile.read(wav_loc)
data1 = data[:10000] / 32768
IPython.display.Audio(data=data1, rate=rate)

```

```python
#Giong gen

wav_loc = "samples/400000-p0_1006_8-vcto-p7.wav"
rate, data = wavfile.read(wav_loc)
data2 = data[:10000] / 32768
IPython.display.Audio(data=data2, rate=rate)
```

```python
# plot
fig, ax = plt.subplots(figsize=(20,4))
ax.plot(data1)
fig, ax = plt.subplots(figsize=(20,4))
ax.plot(data2)
```

```python
# remove noise
data2 = smooth(data2, 5)
data2 = np.array(data2, dtype="float32")
IPython.display.Audio(data=data2, rate=rate)
```
