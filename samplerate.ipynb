```python
from pydub import AudioSegment

path = '/home/topica/StarGAN-Voice-Conversion/sontung/p5'
di = os.listdir(path)

for i in di:
    pa = os.path.join(path, i)
    out = pa.split('.')[0] + '.wav'
    sound = AudioSegment.from_mp3(pa)
    sound.export(out, format="wav")
```
