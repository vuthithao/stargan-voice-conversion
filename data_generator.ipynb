```python
# remove short audio

import wave
import contextlib
import os

path = '/home/topica/Downloads/03/'
wav = os.listdir(path)

for i in wav:
    fname = os.path.join(path, i)
    with contextlib.closing(wave.open(fname,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
    if duration < 1:
        os.remove(fname)
```
